```python
# 파이썬 3.7.1 버전 낮아서 안되는 듯
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
np.random.seed(3)
tf.random.set_seed(3)

Data_set = np.loadtxt('./datasets/ThoraricSurgery.csv', delimiter=',')

X = Data_set[:,0:17]
Y = Data_set[:,17]

model=Sequential()
model.add(Dense(30, input_dim = 17, activation = "relu"))
model.add(Dense(1, activation = "sigmoid"))

# 딥러닝을 실행합니다.
model.compile(loss = "binary_crossentropy", optimizer = "adam", metrics = ["accuracy"])
model.fit(X, Y, epochs = 100, batch_size = 10)
```

```python
X = data = np.linspace(1,2,200)
y = X*4 + np.random.randn(200)

model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(1, input_dim = 1, activation='linear'))
model.compile(optimizer='sgd', loss = 'mse', metrics=['mse'])
model.fit(X,y, batch_size=1, epochs=30)

predict = model.predict(data)
plt.plot(data, predict, 'b', data, y , 'k.')
plt.show()
```

```python
data=[[2,0],[4,0],[6,0],[8.,1],[10,1],[12,1],[14,1]]

X_data=[i[0] for i in data]
y_data=[i[1] for i in data]

plt.scatter(X_data,y_data)
plt.xlim(0,15)
plt.ylim(-.1,1.1)
```

```python
# 기울기 a와 절편 b의 값을 초기화 합니다.
a = 0
b = 0

# 학습률을 정합니다.
lr = 0.05

# 시그모이드 함수를 정의합니다.
def sigmoid(x):
    return 1 / (1 + np.e**(-x))

for i in range(2001):
    for x_data, y_data in data:
        a_diff = x_data *(sigmoid(a*x_data + b) - y_data)
        b_diff = sigmoid(a*x_data + b)-y_data
        a = a - lr * a_diff
        b = b - lr * b_diff
        if i % 1000 == 0: #나머지가 0
            print("epoch=%.f, 기울기=%.04f, 절편=%.04f"%(i,a,b))
```
